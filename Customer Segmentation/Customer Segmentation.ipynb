# Separating Annual Income and Spending Score

```python
x = data.iloc[:, [3,4]].values

print(x)

# x = data['Annual Income (k$)']
# y = data['Spending Score (1-100)']

# print(x)
# print(y)
```

# Choosing the number of clusters

### WCSS > Withing Clusters  Sum of Squares

#### Finding wcss value for different number of clustres

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans( n_clusters = i, init = 'k-means++', random_state = 42 )
    kmeans.fit(x)
    
    wcss.append(kmeans.inertia_)

print(wcss)
```

```python
sns.set()
plt.plot( range(1,11), wcss )
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

# The Optimun Number of Clusters = 5

## Training the Model

```python
model = KMeans( n_clusters = 5, init = 'k-means++', random_state = 0 )


# return a label for each cluster
y = model.fit_predict(x)

print(y)
```

# Visualizing all the Clusters

```python
plt.figure( figsize = (10, 10) )
plt.scatter( x[ y==0, 0], x[ y==0, 1], s=50, c='green', label = 'Cluster 1' )
plt.scatter( x[ y==1, 0], x[ y==1, 1], s=50, c='purple', label = 'Cluster 2' )
plt.scatter( x[ y==2, 0], x[ y==2, 1], s=50, c='gray', label = 'Cluster 3' )
plt.scatter( x[ y==3, 0], x[ y==3, 1], s=50, c='gold', label = 'Cluster 4' )
plt.scatter( x[ y==4, 0], x[ y==4, 1], s=50, c='orange', label = 'Cluster 5' )

plt.scatter( model.cluster_centers_[ :, 0 ], model.cluster_centers_[ :, 1 ], s = 100, c = 'red', label = 'Centroids' )

plt.title( 'Customer Groups' )
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
